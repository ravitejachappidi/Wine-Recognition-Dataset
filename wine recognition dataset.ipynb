## Preprocessing

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
dataset_1 = scaler.fit_transform(dataset)
```

```python
dataset_1 = pd.DataFrame(dataset_1,columns = dataset.columns)
```

```python
dataset_1.head()
```

```python
dataset_1.target.unique()
```

```python
dataset_1.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset_1["target"] = le.fit_transform(dataset_1["target"])
```

```python
from scipy.stats import f_oneway
for i in x:
    statastic,pvalue = f_oneway(dataset_1[i],dataset_1.target)
    print(i,"--",pvalue)
    print("\n")
```

```python
corr_values = {}
from scipy.stats import kendalltau
for i in x:
    corr, _ = kendalltau(dataset_1[i],dataset_1.target)
    print(i,"----",corr)
    print("\n")
    corr_values[i] = corr
```

```python
plt.figure(figsize = (16,6))
sns.heatmap(dataset_1.corr(),annot = True)
plt.show()
```

## Modelling

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
X = dataset_1.drop("target",axis = 1)
y = dataset_1.target
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 6)
```

```python
logReg = LogisticRegression()
logReg.fit(x_train,y_train)
pred_train = logReg.predict(x_train)
pred_test = logReg.predict(x_test)

print("train set accuracy","  ",accuracy_score(y_train,pred_train))
print("test set accuracy","  ",accuracy_score(y_test,pred_test))
print("cross validation score",cross_val_score(logReg,X,y,cv = 20).mean())
```

```python
svc = SVC()
svc.fit(x_train,y_train)
pred_train = svc.predict(x_train)
pred_test = svc.predict(x_test)

print("train set accuracy","  ",accuracy_score(y_train,pred_train))
print("test set accuracy","  ",accuracy_score(y_test,pred_test))
print("cross validation score",cross_val_score(svc,X,y,cv = 20).mean())
```

```python
sgd = SGDClassifier()
sgd.fit(x_train,y_train)
pred_train = sgd.predict(x_train)
pred_test = sgd.predict(x_test)

print("train set accuracy","  ",accuracy_score(y_train,pred_train))
print("test set accuracy","  ",accuracy_score(y_test,pred_test))
print("cross validation score",cross_val_score(sgd,X,y,cv = 20).mean())
```

```python
random_forest = RandomForestClassifier()
random_forest.fit(x_train,y_train)
pred_train = random_forest.predict(x_train)
pred_test = random_forest.predict(x_test)

print("train set accuracy","  ",accuracy_score(y_train,pred_train))
print("test set accuracy","  ",accuracy_score(y_test,pred_test))
print("cross validation score",cross_val_score(random_forest,X,y,cv = 20).mean())
```
